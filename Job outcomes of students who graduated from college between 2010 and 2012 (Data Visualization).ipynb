# Job outcomes of students who graduated from college between 2010 and 2012

All data is from American Community Survey 2010-2012 Public Use Microdata Series and can be downloaded from this website: https://github.com/fivethirtyeight/data/tree/master/college-majors

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

| Columns| Details|
| -------| -------|
| Rank| Rank by median earnings (the dataset is ordered by this column)|
| Major_code| Major code|
| Major| Major description|
| Major_category| Category of major|
| Total| Total number of people with major|
| Sample_size| Sample size (unweighted) of full-time|
| Men| Male graduates|
| Women| Female graduates|
| ShareWomen| Women as share of total|
| Employed| Number employed|
| Median| Median salary of full-time, year-round workers|
| Low_wage_jobs| Number in low-wage service jobs|
| Full_time| Number employed 35 hours or more|
| Part_time| Number employed less than 35 hours|

In this section, we will generate different types of graphs for better understanding the data set.

### 1. Import Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# imported for plots to display inline
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print("The headers of the dataset:")
print(recent_grads.iloc[0])
```

```python
print("The first five rows of the dataset:")
recent_grads.head()
```

```python
print("The last five rows of the dataset:")
recent_grads.tail()
```

```python
print("Statistical information of the dataset:")
recent_grads.describe()
```

```python
raw_data_count = len(recent_grads)
print("Number of rows in this dataset:", raw_data_count)
```

### 2. Data Cleaning

One row contained missing values and have to be dropped.

```python
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = len(recent_grads)
print("Number of rows in this data set after removing missing values:", cleaned_data_count)
```

### 3. Scatter Plots

```python
ax1 = recent_grads.plot(x="Sample_size", y="Median", kind="scatter")
ax1.set_title("Sample_size VS Median")
plt.show(ax1)
```

```python
ax2 = recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", title="Sample_size VS Unemployment_rate")
plt.show(ax2)
```

```python
ax3 = recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Full_time VS Median")
plt.show(ax3)
```

```python
ax4 = recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="ShareWomen VS Unemployment_rate")
plt.show(ax4)
```

```python
ax5 = recent_grads.plot(x="Men", y="Median", kind="scatter", title="Men VS Median")
plt.show(ax5)
```

```python
ax6 = recent_grads.plot(x="Women", y="Median", kind="scatter", title="Women VS Median")
plt.show(ax6)
```

### 4. Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"] 

fig = plt.figure(figsize=(5,12))

for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', bins=10, rot=30)
```
